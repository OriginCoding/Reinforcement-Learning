```python
import gym
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy
import os
```

<h2>2. Loading and Testing Environment

```python
environment_name = 'CarRacing-v0'
env = gym.make(environment_name)
```

```python
env.reset()
```

```python
env.close()
```

```python
env.reset()
```

```python
env.render('human')
```

```python
episodes = 5
for episode in range(1,episodes + 1):
  state = env.reset()
  done = False
  score = 0

  while not done:
    env.render('human')
    action = env.action_space.sample()
    n_state, reward, done, info = env.step(action)
    score += reward
  print('Episode:{} Score:{}'.format(episode,score))
env.close()
```

<h2>3.Train Model

```python
env = gym.make(environment_name)
env = DummyVecEnv([lambda: env])
```

```python
log_path = os.path.join('Training','Logs')
model = PPO('CnnPolicy',env,verbose=1,tensorboard_log=log_path)
```

```python
model.learn(total_timesteps=200000)
```

```python
for training in range(5,21):
    model.learn(total_timesteps=100000)
    model_name = "PPO_DrivingX_" + str(training)
    ppo_path = os.path.join('Training','Saved Models',model_name)
    model.save(ppo_path)
    del model
    model = PPO.load(ppo_path,env)
```

<h2>5. Evaluate and Test

```python
evaluate_policy(model,env,n_eval_episodes=2,render=True)
```

```python
env.close()
```

```python
episodes = 5
for episode in range(1,episodes + 1):
  obs = env.reset()
  done = False
  score = 0

  while not done:
    env.render()
    action, _= model.predict(obs)
    obs, reward, done, info = env.step(action)
    score += reward
  print('Episode:{} Score:{}'.format(episode,score))
env.close()
```
